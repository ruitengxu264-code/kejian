## 为什么要用 Jupyter Notebook？

Jupyter Notebook 的价值，不只是“能跑代码”，而是把下面几件事放在同一个地方：

- 写代码
- 看输出
- 插入解释文字
- 展示图表
- 保留实验过程

这使它特别适合教学、探索式分析、方法演示和研究笔记。

## Notebook 的两个核心单元

### 1. Code Cell

用于运行代码，例如 Python、R、Stata 等。

### 2. Markdown Cell

用于写说明文字、标题、公式、列表和链接。

一个好的 notebook 不是“很多代码块”，而是**代码和解释交替出现**，让别人能看懂你的思路。

## 最常用的运行快捷键

- `Shift+Enter`：运行当前 cell，并跳到下一个 cell
- `Ctrl+Enter`：运行当前 cell，但停留在当前 cell
- `Alt+Enter`：运行当前 cell，并在下方新建一个 cell

这三个快捷键里，最重要的是 `Shift+Enter`。

## Command Mode 和 Edit Mode

Notebook 有两种常见状态：

- `Edit Mode`：正在编辑单元格内容
- `Command Mode`：对整个单元格进行操作

常见切换方式：

- `Enter`：进入编辑模式
- `Esc`：退出编辑模式，进入命令模式

很多快捷键只有在 `Command Mode` 下才有效。

## 常用单元格快捷键（Command Mode）

- `A`：在当前单元格上方插入新 cell
- `B`：在当前单元格下方插入新 cell
- `M`：把当前 cell 改成 Markdown
- `Y`：把当前 cell 改成 Code
- `D` `D`：删除当前 cell
- `Z`：撤销刚刚删除的 cell

这几个已经足够应对大部分课堂和实验场景。

## Quick Commands: 常见 Jupyter / IPython 命令

在 Python notebook 中，有一类非常实用的命令叫做 **magic commands**。它们通常以 `%` 或 `%%` 开头。

````python
%pwd
````

`%pwd` 会显示当前工作目录。很多“为什么找不到文件”的问题，第一步都应该先看这里。

````python
x = 10
y = 20
%whos
````

`%whos` 可以查看当前内存中的变量，非常适合做课堂演示和调试。

````python
%timeit sum(range(1000))
````

`%timeit` 用于快速测试一段代码的运行时间，是非常高频的性能检查工具。

````python
!pwd
!ls
````

前面加 `!` 表示运行 shell 命令。

这在 notebook 里也很常用，比如：

- `!pwd`
- `!ls`
- `!pip list`

不过要注意：`%pwd` 是 IPython magic，`!pwd` 是 shell 命令，它们看起来相似，但来源不同。

## 课堂里最值得记住的几个 quick commands

- `Shift+Enter`：运行当前单元
- `Esc` / `Enter`：切换命令模式与编辑模式
- `A` / `B`：在上方或下方插入单元
- `M` / `Y`：在 Markdown 和 Code 之间切换
- `%pwd`：查看当前目录
- `%whos`：查看变量
- `%timeit`：测试运行时间
- `!pwd`、`!ls`：运行 shell 命令

## 如果 notebook 卡住了怎么办？

常见处理方法：

- 先确认是不是某个 cell 还在运行
- 尝试中断当前 cell
- 必要时重启 kernel
- 重启后从上到下重新运行关键单元

Notebook 不是 Word 文档，它有“状态”。如果 kernel 重启了，变量通常也会消失，所以要学会按顺序重跑。
